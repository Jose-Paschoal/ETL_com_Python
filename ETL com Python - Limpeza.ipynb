# Limpeza dos Dados

```python
import pandas as pd

import pandera as pa
```

```python
df = pd.read_excel('ocorrencia_2010_2020.xlsx')
```

```python
df
```

```python
# experimentando filtro

df.loc[1, 'ocorrencia_cidade']
```

```python
# trazendo todos os dias de uma coluna SEM informar a linha

df.loc[:,'ocorrencia_cidade']
```

```python
# lembrando que o LOC trata de labels, ou seja, pelos nomes das colunas, enquanto que o iloc trata dos ÍNDICES.
```

```python
# Trazendo todos as ocorrências da LINHA UM em todas as colunas

df.loc[1,]
```

```python
# verificando se a coluna código de ocorrência tem valores ÚNICOS

df.codigo_ocorrencia.is_unique
```

```python
# Como vimos que o código não se repete, podemos transformá-lo no índice do dataframe

df.set_index('codigo_ocorrencia', inplace=True)
```

```python
df
```

```python
# Retirando o index que foi inserido há pouco e voltando ao dataframe original

df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
df
```

```python
# Alterando os valores de asterico para NA

df.loc[df.ocorrencia_aerodromo == '****', ['ocorrencia_aerodromo']] = pd.NA
```

```python
df
```

```python
# utilizando o replace para continuar com a limpeza dos dados e como uma forma de acelerar o processo.

df.replace(['**','###!','####','****','*****', 'NULL'], pd.NA, inplace=True)
```

```python
df
```

```python
# checando quantos dados NA há no dataset

df.isna().sum()
```

```python
# criando uma coluna como exemplo e mostrando como deletar

df['codigo_ocorrencia_bkp'] = df.codigo_ocorrencia
```

```python
df
```

```python
df.drop(['codigo_ocorrencia_bkp'], axis=1, inplace=True)
```

```python
df
```

```python
# dropna irá retirar todos os NA do dataframe, ou seja, todas as linhas que contém esses valores.
```

```python
# retirando linhas DUPLICADAS DO DATAFRAME

df.drop_duplicates()
```
