# Transformação de Dados

```python
import pandas as pd

import pandera as pa
```

```python
# importando os dados e já tratando os valores ausentes das aulas passadas

valores_ausentes = ['**','###!','####','****','*****', 'NULL']

df = pd.read_excel('ocorrencia_2010_2020.xlsx', parse_dates=['ocorrencia_dia'], na_values=valores_ausentes)

df.head(10)
```

```python
schema = pa.DataFrameSchema(
    columns = {
        'codigo_ocorrencia': pa.Column(pa.Int),
        'codigo_ocorrencia2': pa.Column(pa.Int),
        'ocorrencia_classificacao': pa.Column(pa.String),
        'ocorrencia_cidade': pa.Column(pa.String),
        'ocorrencia_uf': pa.Column(pa.String, pa.Check.str_length(2,2), nullable=True),
        'ocorrencia_aerodromo': pa.Column(pa.String, nullable=True),
        'ocorrencia_dia': pa.Column(pa.DateTime),
        'ocorrencia_hora': pa.Column(pa.String, nullable=True),
        'total_recomendacoes': pa.Column(pa.Int)
    }
)
```

```python
schema.validate(df)
```

```python
# verificando os tipós de dados

df.dtypes
```

```python
# mesmo resultado se colocar isnull.

df.isna().sum()
```

```python
# localizando o valor nulo através do LOC

df.loc[df.ocorrencia_uf.isnull()]
```

```python
# alterando a semântica

filtro = df.ocorrencia_uf.isnull()

df.loc[filtro]
```

```python
df.count()
```

```python
# fazendo filtro em ocorrências com mais de 10 recomendações

filtro = df.total_recomendacoes > 10
df.loc[filtro]
```

```python
# fazendo filtro em ocorrências com mais de 10 recomendações - outra variação

filtro = df.total_recomendacoes > 10
df.loc[filtro, 'ocorrencia_cidade']
```

```python
# fazendo filtro em ocorrências - outra variação

filtro = df.ocorrencia_classificacao == 'INCIDENTE GRAVE'
df.loc[filtro, 'ocorrencia_cidade']
```

```python
# fazendo filtro em ocorrências - outra variação de filtro

# usando o E...........

filtro1= df.ocorrencia_classificacao == 'INCIDENTE GRAVE'
filtro2= df.ocorrencia_uf == 'SP'

df.loc[filtro1 & filtro2]
```

```python
# fazendo filtro em ocorrências - outra variação de filtro

# usando o OU...........

filtro1= df.ocorrencia_classificacao == 'INCIDENTE GRAVE'
filtro2= df.ocorrencia_uf == 'SP'

df.loc[filtro1 | filtro2]
```

```python
# fazendo filtro em ocorrências - outra variação de filtro

# usando o OU e o E...........

filtro1= (df.ocorrencia_classificacao == 'INCIDENTE GRAVE') | (df.ocorrencia_classificacao == 'INCIDENTE')
filtro2= df.ocorrencia_uf == 'SP'

df.loc[filtro1 & filtro2]
```

```python
# fazendo filtro em ocorrências - outra variação de filtro

# usando o OU e o E...........

filtro1= df.ocorrencia_classificacao.isin(['INCIDENTE GRAVE', 'INCIDENTE'])
filtro2= df.ocorrencia_uf == 'SP'

df.loc[filtro1 & filtro2]
```

```python
# filtros para analisar parte dos valores - ocorrencias com cidades que comecem com a letra C

filtro = df.ocorrencia_cidade.str[0] == 'C'

df.loc[filtro]
```

```python
# filtros para analisar parte dos valores - ocorrencias com cidades que terminem com a letra A

filtro = df.ocorrencia_cidade.str[-1] == 'A'

df.loc[filtro]
```

```python
# filtros para analisar parte dos valores - ocorrencias com cidades que terminem com as letras NA

filtro = df.ocorrencia_cidade.str[-2:] == 'NA'

df.loc[filtro]
```

```python
# filtro para verificar se uma string contrém determinado conjunto de letras

filtro = df.ocorrencia_cidade.str.contains('MA')

df.loc[filtro]
```

```python
# filtro para verificar se uma string contrém determinado conjunto de letras - ADICIONANDO O OR.

# verifica se primeiro nos nomes de cidades há as letras MA ou AL, ou seja, uma ou outra.

filtro = df.ocorrencia_cidade.str.contains('MA|AL')

df.loc[filtro]
```

```python
# filtro para ocorrências do ano de 2015

filtro = df.ocorrencia_dia.dt.year == 2015

df.loc[filtro]
```

```python
# filtro para ocorrências do ano de 2015 e mês 12

filtro1 = df.ocorrencia_dia.dt.year == 2015
filtro2 = df.ocorrencia_dia.dt.month == 12

df.loc[filtro1 & filtro2]
```

```python
# filtro para ocorrências do ano de 2015 e mês 12 - inserindo variações de dias

filtro1 = df.ocorrencia_dia.dt.year == 2015
filtro2 = df.ocorrencia_dia.dt.month == 12
filtro3 = (df.ocorrencia_dia.dt.day >= 3) & (df.ocorrencia_dia.dt.day <=8)

df.loc[filtro1 & filtro2 & filtro 3]
```

```python
# criando nova coluna - unificando dia e hora

# primeiro transformamos a data em string para unir essa coluna com as horas, e depois já convertemos em data.

df['ocorrencia_dia_hora'] = pd.to_datetime(df.ocorrencia_dia.astype(str) + '' + df.ocorrencia_hora)
```

## Agrupamento de Dados

```python
# ocorrencias do anos de 2015 e mês 03 - atribuindo um novo dataframe com dados filtrados

filtro1 = df.ocorrencia_dia.dt.year == 2015
filtro2 = df.ocorrencia_dia.dt.month == 3

df2015 = df.loc[filtro1 & filtro2]
```

```python
df2015
```

```python
df2015.groupby(['ocorrencia_classificacao']).codigo_ocorrencia.count()
```

```python
# filtro mais elaborado para a região SUDESTE - novo dataframe criado

filtro1 = df.ocorrencia_dia.dt.year == 2010
filtro2 = df.ocorrencia_uf.isin(['SP', 'MG', 'ES', 'RJ'])

dfsudeste = df.loc[filtro1 & filtro2]
dfsudeste
```

```python
dfsudeste.groupby(['ocorrencia_classificacao']).size()
```

```python
dfsudeste.groupby(['ocorrencia_classificacao', 'ocorrencia_uf']).size()
```

```python
dfsudeste.groupby(['ocorrencia_cidade']).size().sort_values(ascending=False)
```

```python
filtro = dfsudeste.ocorrencia_cidade == 'RIO DE JANEIRO'
dfsudeste.loc[filtro].total_recomendacoes.sum()
```

```python
# total de recomendações por cidade

dfsudeste.groupby(['ocorrencia_cidade']).total_recomendacoes.sum()
```

```python
# apenas cidades com recomendações

filtro = dfsudeste.total_recomendacoes > 0 
dfsudeste.loc[filtro].groupby(['ocorrencia_cidade']).total_recomendacoes.sum()
```
